```python
# 1. Función básica
def saludar(nombre):
    return f"¡Hola, {nombre}!"

print(saludar("Jonh"))  # ¡Hola, name!
```

```python
# 2. Función con valor por defecto
def potencia(base, exponente = 3):
    return base ** exponente

print(potencia(3))     # 9  (usa exponente=2 por defecto)
print(potencia(3, 3))  # 27
```

```python
# 3. Función con múltiples retornos
def dividir(a, b):
    if b == 0:
        return None, "Error: división entre cero"
    return a / b, "OK"

resultado, estado = dividir(10, 2)
print(resultado, estado)  # 5.0 OK
```

```python
#4. Función con *args (argumentos variables)
def sumar(*numeros):
    print(numeros[0])
    return sum(numeros)

print(sumar(10, 20, 30, 40, 50, 60))  # 10
```

```python
# 5. Función con **kwargs (argumentos con nombre)
def mostrar_info(**datos):
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")

mostrar_info(nombre="Luis", edad=30, ciudad="Lima")
```

```python
# 6. Función lambda (función anónima)
cuadrado = lambda x: x ** 2
print(cuadrado(5))  # 25
```

```python
print("Calculadora Python")
print("Operaciones disponibles: +, -, *, /")

# 1. Pedir operación
operacion = input("Elige una operación (+, -, *, /): ")
num1 = float(input("Ingresa el primer número: "))
num2 = float(input("Ingresa el segundo número: "))

# 2. Opciones Tipo de operaciones
match operacion:
    case '+':
        resultado = num1 + num2
        print(f"Resultado: {num1} + {num2} = {resultado}")
    case '-':
        resultado = num1 - num2
        print(f"Resultado: {num1} - {num2} = {resultado}")
    case '*':
        resultado = num1 * num2
        print(f"Resultado: {num1} * {num2} = {resultado}")
    case '/':
        if num2 == 0:
            print("Error: No se puede dividir entre cero.")
        else:
            resultado = num1 / num2
            print(f"Resultado: {num1} / {num2} = {resultado}")
    case _:
        print("Error: Operación no válida. Por favor elige +, -, * o /.")
```
